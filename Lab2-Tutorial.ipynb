```python
# SciPy Sparse Array
# Creating a CSR matrix from an array
import numpy as np
from scipy.sparse import csr_matrix
arr = np.array([0,0,0,0,0,1,1,0,2])
print(csr_matrix(arr))
```

```python
# Viewing stored data with the data property
arr = np.array([[0,0,0],[0,0,1],[1,0,2]])
print(csr_matrix(arr).data)
```

```python
# Converting csr to csc using tocsc() method
newarr = csr_matrix(arr).tocsc()
print(newarr)
```

```python
# Providing index manually
new_series = pd.Series([5,6,7,8,9,10], index=['a', 'b', 'c', 'd', 'e', 'f'])
print(new_series)
print('----------------------')

# Accessing through custom indexing
print(new_series['f'])

# Retrieving several elements by index and make group assignments
print(new_series[['a','b','c']]==0)

# Filtering and maths operations with pandas
new_series2 = new_series[new_series>7]
print(new_series2)

new_series2 = new_series[new_series>0]*2
print(new_series2)
```
