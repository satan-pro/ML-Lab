```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('company_sales_data.csv')
```

## Question 1

```python
x = np.linspace(0,100, 10)
y = np.sin(x)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title("Question 1")
```

## Question 2

```python
x = np.linspace(0,100,10)
y = np.sin(x)

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.plot(x,y)
ax2.plot(x,y)
plt.show()
```

## Question 3

```python
df = pd.read_csv('company_sales_data.csv')
df.head()
```

```python
x = df['month_number']
y = df['total_profit']

plt.plot(x, y)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Question 3')
plt.grid(True)
plt.show()
```

## Question 4

```python
plt.plot(x,y, 
            linestyle='--', 
             color='red', 
             linewidth=3, 
             marker='o', 
             markersize=8, 
             markerfacecolor='red', 
             label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Question 4')
plt.legend(['Total Profit'], loc='lower right')
plt.grid(True)
plt.show()
```

## Additional 1

```python
x = df['month_number']
columns = df.columns
print(columns)
for column in columns[1:6]:
    plt.plot(x, df[column], marker='o', label=column)
plt.grid(True)
plt.xlabel('Month Number')
plt.ylabel('Plot of different products')
plt.legend()
plt.show()
```

## Additional 2

```python
total_sales = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()

plt.figure(figsize=(10,8))
plt.pie(total_sales, labels = total_sales.index, autopct='%1.1f%%')
plt.title('Total Sales for Each Product Last Year')
plt.legend()
plt.show()
```
